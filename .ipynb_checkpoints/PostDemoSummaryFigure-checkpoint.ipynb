```python
# Import seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Apply the default theme
sns.set_theme()

# Load dataset
data = pd.read_csv("figures\L023.CSV")
data["DateTime"] = pd.to_datetime(data["DateTime"])
```

```python
data.head()
```

```python
# Adjust categorical variable for visualization
upper = np.ma.masked_where(data["ALERT01"]==1,data["Temp"])
lower = np.ma.masked_where(data["ALERT01"]==0,data["Temp"])
```

```python
fig, ax = plt.subplots()
ax.plot(data["DateTime"], upper, data["DateTime"], lower)
plt.xticks(rotation=90, ha='right')
plt.suptitle("June 3rd: Heat-Monitor Demo Data")
```
